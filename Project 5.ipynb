## Analyze traffic accident data to identify patterns related to road conditions, weather, and time of day. Visualize accident hotspots and contributing factors.



Dataset :- https://www.kaggle.com/code/harshalbhamare/us-accident-eda



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/abhin/Downloads/US_Accidents.csv")
```

```python
print(df.head())
```

```python
print(df.tail())
```

```python
print(df.info())
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Timing', order=df['Timing'].value_counts().index, palette='viridis')
plt.title('Accident Distribution by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
road_counts = df['RoadCondition'].value_counts()
sns.barplot(x=road_counts.index, y=road_counts.values, palette='magma')
plt.title('Accident Distribution by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
weather_counts = df['WeatherCondition'].value_counts().head(10)
sns.barplot(x=weather_counts.index, y=weather_counts.values, palette='coolwarm')
plt.title('Top 10 Weather Conditions Leading to Accidents')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
state_accidents = df['State'].value_counts()
sns.barplot(x=state_accidents.index, y=state_accidents.values, palette='rocket')
plt.title('Accident Hotspots by State')
plt.xlabel('State')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=90)
plt.show()
```
